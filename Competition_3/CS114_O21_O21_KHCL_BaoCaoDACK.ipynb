#### Thuật toán học:
1. **XGBoost (Extreme Gradient Boosting)**
* **Mô tả:** XGBoost là một thuật toán học máy mạnh mẽ và tối ưu hóa cho việc sử dụng trong các bài toán hồi quy và phân loại. Nó dựa trên phương pháp tăng cường độ dốc (Gradient Boosting), sử dụng một tập hợp các mô hình yếu (weak learners) để xây dựng một mô hình mạnh hơn.<br/>
* **Đặc điểm nổi bật:**
  * Khả năng xử lý missing values tốt.
  * Hỗ trợ regularization để tránh overfitting.
  * Khả năng parallel processing, giúp tăng tốc độ tính toán.
  * Dễ dàng điều chỉnh các siêu tham số (hyperparameters).
2. **CatBoost (Categorical Boosting)**
* **Mô tả**: CatBoost là một thuật toán học máy dựa trên phương pháp tăng cường độ dốc, được thiết kế đặc biệt để xử lý dữ liệu có biến phân loại (categorical features) một cách hiệu quả.<br/>
* **Đặc điểm nổi bật**:
  * Xử lý tự động các biến phân loại mà không cần chuyển đổi chúng thành số.
  * Giảm nguy cơ overfitting nhờ vào việc sử dụng điều chỉnh và regularization.
  * Hỗ trợ cả dữ liệu phân loại và hồi quy.
  * Dễ dàng sử dụng và cài đặt, với ít yêu cầu về tiền xử lý dữ liệu.
3. **LightGBM (Light Gradient Boosting Machine)**
* **Mô tả**: LightGBM là một framework học máy mạnh mẽ và nhanh chóng dựa trên phương pháp tăng cường độ dốc. Nó được thiết kế để xử lý các tập dữ liệu lớn và phức tạp một cách hiệu quả.<br/>
* **Đặc điểm nổi bật:**
  * Tốc độ học nhanh chóng và hiệu quả bộ nhớ.
  * Sử dụng phương pháp Leaf-wise tree growth (phát triển cây theo lá) thay vì Level-wise tree growth (phát triển cây theo mức), giúp cải thiện hiệu suất.
  * Hỗ trợ tính toán phân tán và parallel learning.
  * Tối ưu hóa cho việc xử lý các dữ liệu lớn và phức tạp.

#### Quá trình huấn luyện:
1. **Chia dữ liệu:** Dữ liệu được chia thành tập huấn luyện và tập kiểm tra sử dụng **train_test_split.**
2. **Chuẩn hóa dữ liệu:** Dữ liệu được chuẩn hóa bằng **StandardScaler** để đảm bảo các đặc trưng có cùng đơn vị đo lường.
3. **Grid Search CV:** Sử dụng GridSearchCV để tìm các tham số tối ưu cho mỗi mô hình. Các tham số khác nhau được thử nghiệm để tìm ra cấu hình tốt nhất dựa trên tiêu chí **R^2**.
4. **Huấn luyện mô hình:** Mô hình được huấn luyện với tập dữ liệu đã chuẩn hóa.

#### Thực nghiệm:
1. **Dự đoán và đánh giá:** Sử dụng tập kiểm tra để dự đoán và tính toán sai số R^2 để đánh giá mô hình.
2. **Lựa chọn mô hình tốt nhất:** So sánh R^2 của các mô hình khác nhau và chọn mô hình có R^2 cao nhất.
3. **Lưu mô hình và scaler:** Mô hình tốt nhất và scaler được lưu lại sử dụng joblib để sử dụng sau này.

```python
# Hàm xử lý tên file
def process_directory(directory_path):
    data = []
    errors = 0
    valid_brands = ["Honda", "Yamaha", "Suzuki", "Others", "VinFast"]
    count_brands = {"Honda" : 0, "Yamaha" : 0, "Suzuki" : 0, "Others" : 0, "VinFast" : 0}
    for brand_folder in os.listdir(directory_path):
        brand_path = os.path.join(directory_path, brand_folder)
        if os.path.isdir(brand_path) and brand_folder in valid_brands:  # Kiểm tra tên thương hiệu hợp lệ
            for filename in os.listdir(brand_path):
                if filename.endswith((".jpg", ".jpeg", ".png", ".webp", ".heic")):

                    # Kiểm tra định dạng tên file
                    parts = filename.split(".")
                    if len(parts) != 4:
                        errors += 1
                        continue  # Bỏ qua file không hợp lệ vì không đủ thành phần

                    count_brands[brand_folder] += 1

    for brand, count in count_brands.items():
        print(f"Tổng số ảnh của thương hiệu {brand}: {count}")

    print(f"Tổng số file không hợp lệ: {errors}")
if __name__ == "__main__":
    df = process_directory(szWorkingDir)
```

### Bước 2 - Data Processing



```python
# Load dữ liệu train và test từ file CSV
def load_train_data(cirSplit):
    df = pd.read_csv(os.path.join(szWorkingDir, "MotocycleDataset-Splits-{}-Train.csv".format(cirSplit)), header=None, names=['path', 'label'])
    return df

def load_test_data(cirSplit):
    df = pd.read_csv(os.path.join(szWorkingDir, "MotocycleDataset-Splits-{}-Test.csv".format(cirSplit)), header=None, names=['path', 'label'])
    return df
```

```python
# Tiền xử lý ảnh
def preprocess_image(img_path):
    try:
      img = load_img(img_path, target_size=(224, 224)) # Kích thước yêu cầu của VGG16
      img = img_to_array(img)
      img = preprocess_input(img) # Chuẩn hóa theo VGG16
      return img
    except (FileNotFoundError, OSError) as e:
      print(f"{e}")
      return None
```

### Bước 3 - Feature Extraction

```python
# Trích xuất đặc trưng bằng VGG16
vgg16 = VGG16(weights='imagenet', include_top=False, pooling = 'avg')

def extract_features(df, output_file):
    with open(output_file, 'w', newline='') as csvfile: # Mở file CSV ở chế độ ghi
        writer = csv.writer(csvfile)

        num_batches = 0
        total_batches = len(df) // 128 + (len(df) % 128 > 0)
        for i in range(0, len(df), 128):
            batch_paths = df['path'][i:i+128]
            batch_images_preprocessed = [preprocess_image(os.path.join(szWorkingDir,img_path)) for img_path in batch_paths]

            # Lọc ra các ảnh có giá trị None ( ảnh không load được)
            batch_images = [img for img in batch_images_preprocessed if img is not None]
            batch_labels = df['label'][i:i + 128].tolist()  # Lấy nhãn của batch tương ứng

            if len(batch_images) == 0:
                print("Bỏ qua batch này vì không load được")
                continue

            batch_images = np.array(batch_images)
            batch_features = vgg16.predict(batch_images, batch_size=128)

            # Chuyển đổi batch_features thành 2D nếu cần
            if len(batch_features.shape) > 2:
                batch_features = batch_features.reshape(batch_features.shape[0], -1)

            # Ghi đặc trưng và nhãn vào file CSV
            for features, label in zip(batch_features, batch_labels):
                writer.writerow(np.append(features, label))

            del batch_images  # Xóa batch_images sau khi đã sử dụng
            num_batches += 1

            print(f"Đã xử lý {num_batches}/{total_batches} batch...")
```

  - SplitID: giá trị từ 1 đến 5

  - Desc: mô tả phương pháp, ví dụ Random. Mỗi nhóm có thể có rất nhiều trials, do đó cần phải mô tả tóm tắt về các trial để còn nhận diện
  


```python
from datetime import datetime
import pytz  # Thêm thư viện pytz để làm việc với múi giờ

def time():
  # Lấy múi giờ Việt Nam
  vietnam_timezone = pytz.timezone('Asia/Ho_Chi_Minh')

  # Lấy thời gian hiện tại theo múi giờ Việt Nam
  now = datetime.now(vietnam_timezone)

  # Định dạng thời gian
  formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")

  # Trả về thời gian thực
  return formatted_date
```

```python
def submit(cirSplit, output_file="/content/drive/MyDrive/ColabNotebooks/Public/CS114.O21-O21.KHCL.ScoreBoard.csv"):
    accuracy = accuracy_splits[cirSplit-1]

    # Dữ liệu cần ghi vào file CSV
    data = {
        'Date': time(),
        'GroupID': '22521626-22520054-22521322',
        'SplitID': cirSplit,
        'Accuracy': f"{accuracy*100:.4f}",  # Định dạng accuracy với 4 chữ số thập phân
        'Desc': "VGG16 feature extraction - SVM"
    }

    # Mở file CSV ở chế độ append (thêm vào cuối file)
    with open(output_file, 'a', newline='') as csvfile:
        fieldnames = ['Date', 'GroupID', 'SplitID', 'Accuracy', 'Desc']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        # Ghi dữ liệu
        writer.writerow(data)
```

```python
for cirSplit in range(1, 6):
  submit(cirSplit)
```
